```python
# cnn model
import matplotlib
import matplotlib.cm as cm
from numpy import mean
from numpy import std
from numpy import dstack
import pandas as pd
from matplotlib import pyplot
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout

from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from matplotlib.collections import LineCollection
```

```python
filenames_train = sorted(listdir("train/Inertial Signals"))
filenames_test = sorted(listdir("test/Inertial Signals"))
```

```python
filenames_train[0][:-10]
```

```python
features = [filenames_train[i][:-10] for i in range(len(filenames_train))]
```

```python
X_train = pd.read_csv("train/Inertial Signals/" + filenames_train[0], header=None, delim_whitespace=True).values
for filename in filenames_train[1:]:
    df = pd.read_csv("train/Inertial Signals/" + filename, header=None, delim_whitespace=True).values
    X_train = dstack([X_train, df])
    
y_train = pd.read_csv("train/y_train.txt", header = None, delim_whitespace=True).values
y_train = y_train - 1
y_train = to_categorical(y_train)
    
X_test = pd.read_csv("test/Inertial Signals/" + filenames_test[0], header=None, delim_whitespace=True).values
for filename in filenames_test[1:]:
    df = pd.read_csv("test/Inertial Signals/" + filename, header=None, delim_whitespace=True).values
    X_test = dstack([X_test, df])
    
y_test = pd.read_csv("test/y_test.txt", header = None, delim_whitespace=True).values
y_test = y_test - 1
y_test = to_categorical(y_test)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
n_timesteps, n_features, n_outputs = X_train.shape[1], X_train.shape[2], y_train.shape[1] # [samples, time steps, features]
```

```python
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_timesteps, n_features)))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Dropout(0.5))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(n_outputs, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs, batch_size = 5, 32
```

```python
%%time
model.fit(X_train, y_train, epochs = epochs, batch_size = batch_size)
```

```python
accuracy = model.evaluate(X_test, y_test, batch_size=batch_size, verbose=0)[1]
```
