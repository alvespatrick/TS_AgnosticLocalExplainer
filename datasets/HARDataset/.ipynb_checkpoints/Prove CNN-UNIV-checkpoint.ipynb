```python
# cnn model
import matplotlib
import matplotlib.cm as cm
from numpy import mean
from numpy import std
from numpy import dstack
import pandas as pd
from matplotlib import pyplot
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout

from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from matplotlib.collections import LineCollection
```

```python
X_train = pd.read_csv("train/X_train.txt", header=None, delim_whitespace=True).values
y_train = pd.read_csv("train/y_train.txt", header = None, delim_whitespace=True).values
y_train = y_train - 1
y_train = to_categorical(y_train)

X_test = pd.read_csv("test/X_test.txt", header=None, delim_whitespace=True).values  
y_test = pd.read_csv("test/y_test.txt", header = None, delim_whitespace=True).values
y_test = y_test - 1
y_test = to_categorical(y_test)

n_features = 1
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], n_features))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], n_features))
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
n_timesteps, n_outputs = X_train.shape[1], y_train.shape[1] # [samples, time steps, features]
```

```python
n_timesteps, n_outputs, n_features
```

```python
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_timesteps, n_features)))
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(Dropout(0.5))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(n_outputs, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs, batch_size = 5, 64
```

```python
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size)
```

```python
accuracy = model.evaluate(X_test, y_test, batch_size=batch_size, verbose=0)[1]
```

```python
accuracy
```

```python
import shap
```

```python
explainer = shap.DeepExplainer(model, X_train[:100])
```

```python
%%time
# explain the first 10 predictions
# explaining each prediction requires 2 * background dataset size runs
shap_values = explainer.shap_values(X_test[:10])
```

```python
class_n = len(shap_values)
class_names = ["WALKING", "WALKING_UPSTAIRS", "WALKING_DOWNSTAIRS", "SITTING", "STANDING", "LAYING"]
row = 6
colors_list = []
mapper_list = []
inpt, outpt = 0, X_train.shape[1]
#inpt, outpt = 0, 100
for class_index in range(class_n):
    flat_shap = np.ravel(shap_values[class_index][row])#shap_norm.ravel()#np.log(shap_norm.ravel()+1)
    lst = list(flat_shap)

    minima = min(lst)
    maxima = max(lst)

    norm = matplotlib.colors.DivergingNorm(vmin=minima, vcenter=0, vmax=maxima)
    mapper = cm.ScalarMappable(norm=norm, cmap=cm.coolwarm)
    mapper_list.append(mapper)
    colors = []
    for v in lst:
        colors.append(mapper.to_rgba(v))
    colors_list.append(colors)



a = X_train[row].ravel()
b = list(range(len(a)))
segm_list = []
for i, shapley_n in enumerate(a):
    if (i != 0) and (i != len(a) - 1):
        next_shapley = a[i+1]
        prev_shapley = a[i-1]
        segm = [(i - 0.5, (prev_shapley + shapley_n)/2),(i, shapley_n),(i + 0.5, (next_shapley + shapley_n)/2)]
    elif i == 0:
        next_shapley = a[i+1]
        segm = [(0, shapley_n), (i + 0.5, (next_shapley + shapley_n)/2)]
    else:
        prev_shapley = a[i-1]
        segm = [(i - 0.5, (prev_shapley + shapley_n)/2), (len(a)-1, shapley_n)]
    segm_list.append(segm)
    

#x = list(range(X_train.shape[1]))
#y = X_train[row].ravel()[:finoa]

#c = colors[:finoa]  
for i, colors in enumerate(colors_list):
    lc = LineCollection(segm_list[inpt:outpt], colors=colors[inpt:outpt], linewidths=2)
    fig, ax = plt.subplots(figsize=(20, 8))
    ax.add_collection(lc)
    #ax.add_collection(X_train[0].ravel()[:finoa])
    ax.autoscale()
    ax.margins(0.1)
    plt.title(class_names[i])
    plt.colorbar(mapper_list[i])
    plt.show()

#plt.show()
```
