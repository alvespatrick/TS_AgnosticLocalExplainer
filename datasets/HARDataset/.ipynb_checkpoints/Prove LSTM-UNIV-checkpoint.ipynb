```python
# cnn model
import matplotlib
import matplotlib.cm as cm
from numpy import mean
from numpy import std
from numpy import dstack
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import LSTM
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
from keras.models import load_model
```

```python
X_train = pd.read_csv("train/X_train.txt", header=None, delim_whitespace=True).values
y_train = pd.read_csv("train/y_train.txt", header = None, delim_whitespace=True).values
y_train = y_train - 1
y_train = to_categorical(y_train)

X_test = pd.read_csv("test/X_test.txt", header=None, delim_whitespace=True).values  
y_test = pd.read_csv("test/y_test.txt", header = None, delim_whitespace=True).values
y_test = y_test - 1
y_test = to_categorical(y_test)

n_features = 1
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], n_features))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], n_features))
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
n_timesteps, n_outputs = X_train.shape[1], y_train.shape[1] # [samples, time steps, features]
```

```python
n_timesteps, n_outputs, n_features
```

```python
model = Sequential()
model.add(LSTM(100))
model.add(Dense(n_outputs, activation='softmax'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
epochs, batch_size = 3, 32
```

```python
%%time
model.fit(X_train, y_train, epochs = epochs, batch_size = batch_size)
model.save('lstm-univ.h5')  # creates a HDF5 file 'my_model.h5'
#del model  # deletes the existing model
```

```python
model = load_model('lstm-univ.h5')
```

```python
accuracy = model.evaluate(X_test, y_test, batch_size=batch_size, verbose=0)[1]
```
